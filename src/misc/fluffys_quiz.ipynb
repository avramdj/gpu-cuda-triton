```python
import torch
import torch.nn.functional as F
from torch.optim import AdamW
```

```python
class Norm(torch.nn.Module):
    def __init__(self, dim, eps=1e-5):
        super().__init__()
        self.weight = torch.nn.Parameter(torch.zeros(dim))
        self.eps = eps

    def forward(self, x):
        return F.rms_norm(x, normalized_shape=x.shape, weight=self.weight + 1, eps=self.eps)
```

```python
lr, wd, dim = 1e-3, 1, 128
```

```python
norm = Norm(dim)
opt = AdamW([norm.weight], lr=lr, weight_decay=wd)
```

```python
print(torch.norm(norm.weight.data, p=2)) # 0.0
for i in range(10000):
    x = torch.randn(dim)
    loss = F.mse_loss(norm(x), torch.randn(dim))
    loss.backward()
    opt.step()
    opt.zero_grad()

print(torch.norm(norm.weight.data, p=2)) # ?
```
